```python
# Adult dog size could be a feature people care about when adopting, so I'm adding it in
# This data comes from AKC Breed Info
dogSize = pd.read_csv('AKC Breed Info.csv', na_values='na', encoding='latin')
dogSize.head()
```

```python
dogSize = dogSize[['Breed', 'size']]
dogSize.head()
```

```python
# I want a dictionary of breed:size category in order to map it to the breeds and mixes in my data
#https://stackoverflow.com/questions/26716616/convert-a-pandas-dataframe-to-a-dictionary
dogSizeDict = dogSize.set_index('Breed').T.to_dict('list')
```

```python
#dogSizeDict
```

```python
def like_function(x):
    group = "unknown"
    for key in dogSizeDict:
        if key in x:
            group = dogSizeDict[key][0]
            break
    return group

adoptedDogs['size'] = adoptedDogs['Breed'].apply(like_function)
```

```python
adoptedDogs.head()
```

```python
adoptedDogs['size'].value_counts()
```

```python
# Checking on the breeds that are sized "unknown"
#adoptedDogs[adoptedDogs['size']=='unknown']['Breed'].value_counts()
```

```python
# Creating dummies for size
sizeDummies = pd.get_dummies(adoptedDogs['size'], drop_first=True)
```

```python
adoptedDogs = pd.concat([adoptedDogs, sizeDummies], axis=1)
adoptedDogs.head()
```

```python
# Looking at Colors for dogs (commented out b/c large output)
#adoptedDogs['Color'].value_counts()
```

```python
#colors: black, white/cream, brown/chocolate, tan/buff/yellow, red, sable, blue/gray, multicolor(inc. tricolor)
#https://stackoverflow.com/questions/26577516/pandas-test-if-string-contains-one-of-the-substrings-in-a-list
adoptedDogs['black'] = np.where(adoptedDogs['Color'].str.contains('Black'), 1, 0)
adoptedDogs['white'] = np.where(adoptedDogs['Color'].str.contains('White|Cream'), 1, 0)
adoptedDogs['brown'] = np.where(adoptedDogs['Color'].str.contains('Brown|Chocolate'), 1, 0)
adoptedDogs['tan'] = np.where(adoptedDogs['Color'].str.contains('Tan|Buff|Yellow'), 1, 0)
adoptedDogs['red'] = np.where(adoptedDogs['Color'].str.contains('Red'), 1, 0)
adoptedDogs['sable'] = np.where(adoptedDogs['Color'].str.contains('Sable'), 1, 0)
adoptedDogs['blue'] = np.where(adoptedDogs['Color'].str.contains('Blue|Gray'), 1, 0)
adoptedDogs['multicolor'] = np.where(adoptedDogs['Color'].str.contains('Tricolor| /'), 1, 0)
adoptedDogs.head()
```

```python
# Creating a flag for whether or not the dog is a mix or purbred
adoptedDogs['mix_breed_label'] = np.where(adoptedDogs['Breed'].str.contains('Mix'), 1, 0)
adoptedDogs['mix_breed_label'].value_counts()
```

Unsurprisingly, most dogs adopted from the shelter are mixes. But there are a decent number of purebreds (or dogs with full parentage known).

```python
# adding this one flag is probably not sufficient to cover all the interesting aspects of a breed
# but I don't have more information readily available at the moment
# would be interesting to do image analysis on breeds here to pull important distinguishing characteristics
blackListedDogs = ['Pit Bull', 'Staffordshire', 'Rottweiler', 'German Shepherd', 'Presa Canario', 'Mastiff', 'Chow Chow', 'Doberman', 'Akita', 'Cane Corso', 'Great Dane', 'Alaskan Malamute', 'Siberian Husky']
adoptedDogs['blacklisted_breed'] = np.where(adoptedDogs['Breed'].str.contains('|'.join(blackListedDogs)),1,0)
adoptedDogs.head()
```

```python
typeDummies = pd.get_dummies(adoptedDogs['Intake Type'], drop_first=True)
condDummies = pd.get_dummies(adoptedDogs['Intake Condition'], drop_first=True)
sexDummies = pd.get_dummies(adoptedDogs['Sex upon Outcome'], drop_first=True)
```

```python
adoptedDogs = pd.concat([adoptedDogs, typeDummies, condDummies, sexDummies], axis=1)
```

```python
# modeling time
y = adoptedDogs['Time_to_Outcome'] #in days
y.describe()
```

```python
X = adoptedDogs.drop(['Time_to_Outcome'], axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)
regr.score(X_test, y_test)
```

```python
lasso = linear_model.Lasso()
lasso.fit(X_train, y_train)
lasso.score(X_test, y_test)
```

```python
ridge = linear_model.Ridge()
ridge.fit(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
gbr.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
rfr.score(X_test, y_test)
```

```python
#Of those models with default settings, the random forest regressor is performing the best. 
#I'm going to check the feature importances to understand what mattered the most
```
